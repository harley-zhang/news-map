```python
import json
from shapely.geometry import shape, box
from shapely.ops import transform
import pyproj
```

```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Define the border box
southwest = (40.477399, -74.259090)
northeast = (40.917577, -73.700272)
bbox = box(southwest[1], southwest[0], northeast[1], northeast[0])
```

```python
# Define file paths
input_file_path = './drive/MyDrive/Colab Notebooks/national_zcta_2020.json'
output_file_path = './drive/MyDrive/Colab Notebooks/filtered_zcta_2020.json'
```

```python
# Read the JSON data
with open(input_file_path, 'r') as file:
    data = json.load(file)
```

```python
# Initialize filtered features list
filtered_features = []

# Filter features
for feature in data['features']:
    geom = shape(feature['geometry'])
    if bbox.contains(geom):
        filtered_features.append(feature)
```

```python
# Create new JSON data
filtered_data = {
    "type": "FeatureCollection",
    "features": filtered_features
}
```

```python
# Save filtered data to a new JSON file
with open(output_file_path, 'w') as file:
    json.dump(filtered_data, file, indent=2)
```
